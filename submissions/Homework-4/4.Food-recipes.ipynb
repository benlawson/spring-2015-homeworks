```python
def print_hierarcy(X):
#hieracrhy clustering code as seen in class
    Z = hr.linkage(X, method='complete', metric='euclidean')
#let's see what this look like (code from lecture notes)
    fig = plt.figure(figsize=(25,10))
    T = hr.dendrogram(Z,color_threshold=4, leaf_font_size=4)
    fig.show()
```

This fun function calculates the average commonality of each ingrident based on its regions. For instance, for the first large recipe, the first 'chinese' entry listed below, I summed up the number of times each of it's ingridents appeared in other recipes from the 'chinese' region. I divided that sum by the region size and then summed the rest of the ingridents in this fashion. Then I divided by the number of ingridents and multiplied by 100 to get a percent. As we can see, all of these recipes have very uncommon averages for their ingrident frequncies. 


```python
temp_list1 = []
#temp_list2 = []
for idx, row in lots_of_ing.iterrows():
    a = 0
    regional_size = region_dict[row.regions].count()[0] #number of recipes in region
    for ing in copy.iloc[idx,:]: 
         if (ing != 0):
                a = region_dict[row.regions][ing].sum()/regional_size/row.num_ing*100
    print idx, row.regions, "\t", (a)
    #temp_list.append(idx)
    temp_list1.append(a)
#lots_freq = pd.DataFrame(temp_list2, temp_list)    
```

This just compares 20 of the smallest recipes against the 20 largest recipes. This regrettable attempt at a graphs somewhat helps show that smaller recipes have more in common with their respective regional clusters because the blue bars generally are above 25% while the green bars (large recipes) stay below 3%.

Let's check this against the entire dataset. If the recipe is less than 19 ingridents and has less than 3% similar common ingridents it will not support the hypothsis. If the recipe is less than 19 ingridents but has more than 3% similar common ingridents it will support the hypothsis (but will not be counted because they don't prove anything about the hypothesis). If the recipe is 19 or more ingridients and is less than 3% similar it will support the hypothsis. We have already seen that there are not recipes with 19 or more and greater than 3% similarity. 

```python
false = 0 #these variables will be used to see how much data supports or doesn't support the hypothesis
true = 0 #that large recipes are not similar to other recipes because of their size
for idx, row in binary.iterrows():
    a = 0
    regional_size = region_dict[row.regions].count()[0] #number of recipes in region
    for ing in copy.iloc[idx,:]: 
        if (ing != 0):
            a = region_dict[row.regions][ing].sum()
    a = a/regional_size/row.num_ing*100
    if (a<3): #using 3% similar as the threshold
        if (row.num_ing < 19):
            #print idx, "  do not support hypothesis"
            false = false + 1
        else:
            #print idx, "   support hypothesis"
            true = true + 1
                
print "supporting cases: ", true
print "not supporting cases: ", false
#print japanese.banana.sum()
```
