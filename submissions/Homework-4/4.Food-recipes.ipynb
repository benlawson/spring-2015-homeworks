```python
#read data into dataframe
#I parsed it first in excel
# I chose to use the menu_reciples because it was a smaller dataset and my computer is a bit slow at times
df = pd.read_csv('.\scirep-cuisines-detail\\menu_recipes.csv', index_col=False)
print df.info()
print df.head(10)
```

```python
#change columns names to something more appropiate 
new_names = ['region']
for i in range(0,25):
    new_names.append('ingredient_' + str(i))
print new_names
df.columns = new_names
print df.info()
```

```python
df.fillna(value=0, inplace=True)
print df.head(5)
```

```python
#vectorize the categorical data. Each column is an ingrident and each row is a recipe
#1's represent that that ingriedent is in the recipe
#a '0' column was used to accommedate NaN's
regions = df.region #save regions for later use
df.drop(['region'],axis=1, inplace=True) #drop the region column because we will try to predict this later
binary = pd.DataFrame()
for column in df.columns:
    dummy = pd.get_dummies(df[column])
    binary = binary.add(dummy, fill_value=0)
binary.drop([0],axis=1, inplace=True) #drop the column stating that ingriedent is not including in recipe

```

```python
#manually check that all ingredients are in the row
pd.set_option('display.max_columns', None)
print binary.head(1)
pd.reset_option('display.max_columns')
```

```python
#calculate the number of ingriedents per recipe 
binary['num_ing'] = binary.sum(axis=1)
binary.head(10)
```

```python
binary['regions'] = regions
asian = binary[binary.regions == ('asian')]
chinese = binary[binary.regions == 'chinese']
east_asian = binary[binary.regions == 'east_asian']
italian = binary[binary.regions == 'italian']
japanese = binary[binary.regions == 'japanese']
korean = binary[binary.regions == 'korean']
mexico = binary[binary.regions == 'mexico']
western = binary[binary.regions == 'western']

chinese.drop(['regions'],axis=1, inplace=True)
asian.drop(['regions'],axis=1, inplace=True) 
east_asian.drop(['regions'],axis=1, inplace=True) 
italian.drop(['regions'],axis=1, inplace=True) 
japanese.drop(['regions'],axis=1, inplace=True) 
korean.drop(['regions'],axis=1, inplace=True) 
mexico.drop(['regions'],axis=1, inplace=True) 
western.drop(['regions'],axis=1, inplace=True)

```

```python
def print_hierarcy(X):
#hieracrhy clustering code as seen in class
    Z = hr.linkage(X, method='complete', metric='euclidean')
#let's see what this look like (code from lecture notes)
    fig = plt.figure(figsize=(25,10))
    T = hr.dendrogram(Z,color_threshold=4, leaf_font_size=4)
    fig.show()
```

```python
#sorry not very experienced with matlab/plotting can't figure out how to label the individual plots
#but this visualization is interesting: it runs through all the ingridients for regions by regions
#so I am going to take a closer look at the forth regional detrograph, japanese, and examine the one
#recipe that is alone in its cluster at the threshold of 4
regions = [asian, chinese, east_asian,italian,japanese,korean,mexico,western]
for region in regions:
    print_hierarcy(region) 
```
