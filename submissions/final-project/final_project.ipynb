```python
#get all the entries for Spotify
#parameters - a beautiful soup object (parsed html)
#returns two lists, the song and the time it was listened to
def get_songs_dates(soup):
    dates = []
    songs = []         
    for tag in soup.find_all('div'):
        try:
            if "contents" in tag['class']:
                #print tag.strings
                numbered = tag.strings
        except:
              continue 
    numbered = list(numbered)
    print len(numbered)
    for string in numbered:
        if "2015" in string or "2014" in string or "2013" in string or "2012" in string:
            if " listened " in numbered[numbered.index(string)+ 1]:
                dates.append(string)
                songs.append(numbered[numbered.index(string)+ 1])
    return songs, dates
    
```

```python
s1, d1 = get_songs_dates(soup_tan)
s2, d2 = get_songs_dates(soup_tsui)
s3, d3 = get_songs_dates(soup_gadre)
s4, d4 = get_songs_dates(soup_lawson)
```

```python
def clean_song(row):
    """cleans data by removing 'name..listen to' and splits into two
    returns list of size two"""
    string = row.songs
    string = string.split(" listened to ")[1]
    song = string.split(' by ')[0].replace(' ','_')
    try:
        artist = string.split(' by ')[1].split('on Spotify.')[0].replace(' ','_')
    except:
        #this is the case when it is an 'introduction' to a playlist (aka dumb)
        song = string.split(' - ')[1].replace(' ','_')
        artist = string.split(' - ')[0].split('on Spotify.')[0].replace(' ','_')
    return song
```

```python
def clean_artist(row):
    """cleans data by removing ben lawson and splits into two
    returns list of size two"""
    string = row.songs
    string = string.split(" listened to ")[1]
    #song = string.split(' by ')[0].replace(' ','_')
    try:
        artist = string.split(' by ')[1].split('on Spotify.')[0].replace(' ','_')[:-1]
    except:
        #this is the case when it is an 'introduction' to a playlist (aka dumb)
        #song = string.split(' - ')[1].replace(' ','_')
        artist = string.split(' - ')[0].split('on Spotify.')[0].replace(' ','_')
    return artist
```

```python
#REUSE THIS BLOCK to make graphs for each participant. This should have become a function
#Changed 'music_data' dataframe to each of the partipants corresponding dataframe (2,3,4)

week = music_data[music_data.day < datetime(2015,2,1) ]
#week = week.reindex(index=range(len(week)))
week = week[week.day > datetime(2015,1,1)]
#week = week.reindex(index=range(len(week)))
print week.info()
```

```python
#This is the distance function: if the time are the same it will return ZERO because this is an error in the data. 
#Parameters: two rows of DataFrame
#Returns: three variables in the format of an edge list: "node" "node" "edge weight"
def time_distance(artistA, artistB):
    frame1 = week.iloc[artistA]
    frame2 = week.iloc[artistB]
    timeA = frame1.full_date
    timeB = frame2.full_date
    try:
        return frame1.artist, frame2.artist, (1/abs((timeB-timeA).total_seconds()/60.0))
    except: 
        return frame1.artist, frame2.artist, (0)
```

```python
start = timeit.default_timer()

#df = pd.DataFrame(columns=('source', 'destination', 'genre','style','mood'))
df = []
ind = 0

for i in range(len(week)):
    df2 = [(0,0,0)] * 20
    if i < len(week)-20:
        for j in range (i,i+20):
            if i != j:
                a,b,c = time_distance(i,j)
                df2[(i-j)] = a,b,c 
    else:
        for j in range(i,len(week)):
             if i != j:
                a,b,c = time_distance(i,j)
                df2[i-j] = a,b,c 
    for x in df2:
        df.append(x)
stop = timeit.default_timer()
print (stop - start)
```

```python
full = full.sort('Rainfall Per Day').fillna(0)
#print full.head(10)
grouped = full.groupby('Rainfall Per Day').sum()
percent = grouped
percent['sumz'] = percent.sum(axis=1)
totals = percent.sum(axis=0)
#print percent['Pop/Rock'].apply(lambda d: d/(percent['Pop/Rock'].sum(axis=0)))

#normalize in respect to frequency for genre
for column in percent.columns:
    percent[column] = percent[column].apply(lambda d: float(d)/totals[column])
percent['sumz'] = percent.sum(axis=1)
percent = percent[percent.sumz > 0]

#normalize in repsect for total songs
percent['Pop/Rock'] = percent['Pop/Rock']/percent.sumz
percent['Electronic'] = percent['Electronic']/percent.sumz
percent['Rap'] = percent['Rap']/percent.sumz
percent['R&B'] = percent['R&B']/percent.sumz
percent['Country'] = percent['Country']/percent.sumz
percent['Jazz'] = percent['Jazz']/percent.sumz
percent['Classical'] = percent['Classical']/percent.sumz
percent['Folk'] = percent['Folk']/percent.sumz
percent['International'] = percent['International']/percent.sumz
percent['Stage_&_Screen'] = percent['Stage_&_Screen']/percent.sumz
percent['Vocal'] = percent['Vocal']/percent.sumz
percent.drop('sumz',axis=1,inplace=True)
```

```python
#taken from stack overflow
#This rounds the bins up to the nearest 5 (make it look pretty)
import math
def roundup(x):
    return int(math.ceil(x / 5.0)) * 5
```

```python
# Bin the data frame by "a" with 10 bins...

bins = np.linspace(percent.index.min(), percent.index.max(), 10)
bin2 = [0.]
for x in bins:
    bin2.append(x)
bin2[1] = bin2[1]+3
bin2 = [roundup(x) for x in bin2]
groups = percent.groupby(pd.cut(percent.index, bin2)).sum()
```

```python
#normalize in repsect for total songs
groups['sumz'] = groups.sum(axis=1)

groups['Pop/Rock'] = groups['Pop/Rock']/groups.sumz
groups['Electronic'] = groups['Electronic']/groups.sumz
groups['Rap'] = groups['Rap']/groups.sumz
groups['R&B'] = groups['R&B']/groups.sumz
groups['Country'] = groups['Country']/groups.sumz
groups['Jazz'] = groups['Jazz']/groups.sumz
groups['Classical'] = groups['Classical']/groups.sumz
groups['Folk'] = groups['Folk']/groups.sumz
groups['International'] = groups['International']/groups.sumz
groups['Stage_&_Screen'] = groups['Stage_&_Screen']/groups.sumz
groups['Vocal'] = groups['Vocal']/groups.sumz
groups.drop('sumz',axis=1,inplace=True)
```
